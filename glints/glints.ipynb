```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd

keyword = "teknik"
r = Request("https://glints.com/id/opportunities/jobs/explore?keyword=" + keyword +"&country=ID&locationName=Indonesia", headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)'})
print(r)
response = urlopen(r).read()
print(type(response))

soup = BeautifulSoup(response, "lxml")
print(type(soup))

# inisialisasi array untuk menampilkan hasil
title_lowongan = []
nama_perusahaan = []
lokasi_perusahaan = []
link_lowongan = []
deskripsi_lowongan = []

jobList = soup.find_all("div","compact_job_card")
for p in jobList:
    lowongan = p.find("h3", "CompactOpportunityCardsc__JobTitle-sc-1xtox99-7").get_text()
    # menggunakan try except karena ada beberapa perusahaan yang dirahasiakan
    try:
        perusahaan = p.find('a', "CompactOpportunityCardsc__CompanyLink-sc-1xtox99-8 bFxfxR").get_text()
    except:
        perusahaan = "Perusahaan Dirahasiakan"
    lokasi = p.find("div", "CompactOpportunityCardsc__OpportunityInfo-sc-1xtox99-13 hiAtMr").get_text()
    link = "https://glints.com"+p.find('a').get('href')
    
    #   scrape detail sesuai link
    r2 = Request(link, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)'})
    print(r2)
    response2 = urlopen(r2).read()
    print(type(response2))
    
    soup2 = BeautifulSoup(response2, "html.parser")
    print(type(soup2))
    raw_deskripsi = soup2.find('div', "Opportunitysc__JobDescriptionContainer-sc-1gsvee3-6 gBmrdW")
    # increment deskripsi
    deskripsi = ""

    if raw_deskripsi is None:
        print("raw_deskripsi Nonetype. link: "+ link)
    else:
        for string in raw_deskripsi.strings:
            new_string  = deskripsi + " " + string
            deskripsi = new_string
    
    title_lowongan.append(lowongan)
    nama_perusahaan.append(perusahaan)
    lokasi_perusahaan.append(lokasi)
    deskripsi_lowongan.append(deskripsi.replace("Informasi Penting Pastikan perusahaan yang kamu lamar resmi dengan memeriksa website dan lowongan kerja mereka. Read Less ", ""))
    link_lowongan.append(link)

jobList_dict ={'lowongan':title_lowongan, 'perusahaan':nama_perusahaan, 'lokasi':lokasi_perusahaan, 'deskripsi':deskripsi_lowongan, 'link':link_lowongan}
df = pd.DataFrame(jobList_dict,columns = ['lowongan','perusahaan','lokasi', 'deskripsi', 'link'])

df.sort_values('lowongan',ascending=True)
```
